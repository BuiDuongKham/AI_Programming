### Step 1. Import the necessary libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv).

```python
url = 'https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv'
```

### Step 3. Assign it to a variable called tips

```python
tips = pd.read_csv(url)
tips.head()
```

### Step 4. Delete the Unnamed 0 column

```python
# delete column unnamed
tips = tips.drop('Unnamed: 0', axis=1)
tips.head()
```

### Step 5. Plot the total_bill column histogram

```python
#plot total bill column histogram
sns.displot(tips['total_bill'])
plt.show()
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.show()
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
sns.pairplot(tips)
plt.show()
```

### Step 8. Present the relationship between days and total_bill value

```python
# relationship between days and total bill value
sns.stripplot(x='day', y='total_bill', data=tips)
plt.show()
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
# Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex
sns.scatterplot(x='day',y='tip',data = tips, hue='sex')
plt.show()
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='time')
plt.show()
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
# Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
sns.histplot(ax=axes[0], data=tips[tips['time'] == 'Dinner'], x='tip')
axes[0].set_title('Dinner')
sns.histplot(ax=axes[1], data=tips[tips['time'] == 'Lunch'], x='tip')
axes[1].set_title('Lunch')
plt.show()
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
male_tips = tips[tips['sex'] == 'Male']
female_tips = tips[tips['sex'] == 'Female']
sns.scatterplot(data = male_tips , x = 'total_bill', y = 'tip', hue = 'smoker', ax = axes[0])
axes[0].set_title("Male")
sns.scatterplot(data = female_tips , x = 'total_bill', y = 'tip', hue = 'smoker', ax = axes[1])
axes[1].set_title("Female")
plt.show()
```
