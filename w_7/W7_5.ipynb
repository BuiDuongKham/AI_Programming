### Step 3. Assign it to a variable titanic 

```python
titanic = pd.read_csv("https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/titanic_train.csv")
titanic.head()
```

### Step 4. Set PassengerId as the index 

```python
titanic.set_index('PassengerId')
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
# Đếm số hành khách nam / nữ
titanic_gender = titanic.groupby('Sex').count()
str_format = "%.2f%%"
plt.pie(titanic_gender['PassengerId'], autopct=str_format,labels=['Female','Male'])
plt.title('Female vs Male')
plt.show()
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
titanic['Gender'] = [ 1 if i == "female" else 0 for i in titanic['Sex'] ]
plt.scatter(titanic['Fare'], titanic['Age'], c = titanic['Gender'])
legend_labels = {'0': 'Male', '1': 'Female'}
handles = [plt.Line2D([], [], marker='o', color='w', label=label, markerfacecolor=color) for color, label in legend_labels.items()]
plt.xlabel('Fare')
plt.ylabel('Age')
plt.title('Fare vs Age')
plt.legend(handles=handles)
plt.show()
```
