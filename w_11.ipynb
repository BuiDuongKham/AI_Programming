# Visualizing Company Sale data

### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://github.com/thieu1995/csv-files/blob/main/data/matplotlib/company_sales_data.csv).

```python
df = pd.read_csv("https://raw.githubusercontent.com/thieu1995/csv-files/main/data/matplotlib/company_sales_data.csv")
df.head()
```

### Exercise 1: Read Total profit of all months and show it using a line plot

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/01.png?raw=1)

```python
fig, ax =plt.subplots(figsize=(20,5))
ax.plot(df.month_number,df.total_profit)
ax.set_xlabel("Month number")
ax.set_ylabel("Total profit")
ax.set_title("Company Profit per Month")
plt.show()
```

### Exercise 2: Get total profit of all months and show line plot with the following Style properties

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/02.png?raw=1)

```python
fig, ax =plt.subplots(figsize=(20,5))
ax.plot(df.month_number,df.total_profit, color = 'r' , marker= 'o',markeredgecolor = 'r',markerfacecolor='k', linestyle='--', label = 'profit data of last year')
ax.set_xlabel("Month number")
ax.set_ylabel("Total profit")
ax.set_xticks(df.month_number)
ax.set_title("Company Profit per Month")
plt.legend(loc = 'lower right')
plt.show()
```

### Exercise 3: Read all product sales data and show it  using a multiline plot

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/03.png?raw=1)

```python
fig, ax =plt.subplots(figsize=(20,5))
for i in range(1,len(df.columns)-1):
  ax.plot(df.month_number,df[str(df.columns[i])], label = f'{str(df.columns[i])} sales data' , marker = 'o')
plt.legend()
plt.title('Sales Data')
ax.set_xticks(df.month_number)
plt.show()
```

### Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot
![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/04.png?raw=1)

```python
fig,ax =plt.subplots()
ax.scatter(df.month_number,df.toothpaste, label = 'ToothPaste sales Data')
ax.grid(True, axis= 'both',linestyle='--' )
plt.legend()
plt.show()
```

### Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/05.png?raw=1)

```python
width = 0.35
fig,ax = plt.subplots()
ax.bar(df.month_number-width/2, df.facecream, width=width , label = "Face Cream sales Data")
ax.bar(df.month_number+width/2, df.facewash, width=width , label = "Face Cream sales Data")
ax.set_xticks(df.month_number)
ax.grid(True, linestyle='--')
ax.set_xlabel('Month Number')
ax.set_ylabel('Sale Unit in number')
plt.title(f'Face Wash And Face Cream Sale')
plt.legend(loc='upper left')
plt.show()
```

### Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/07.png?raw=1)

```python
import matplotlib.ticker as ticker
fig, ax = plt.subplots()
ax.hist(df.total_profit, label= "Profit Data")
ax.yaxis.set_major_locator(ticker.MaxNLocator(integer=True))
ax.set_xlabel('Profit range in dollar')
ax.set_ylabel('Actual Profit in Dollar')
plt.legend(loc='upper left')
plt.show()
```

### Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/08.png?raw=1)

```python
fig, ax = plt.subplots()
labels = df.columns[1:-2]
values = df.loc[0][df.columns[1:-2]]
format = '%.1f%%'
ax.pie(values, labels = labels, autopct=format)
ax.set_title('Sales Data')
plt.legend(loc='lower right')
plt.show()
```

### Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/09.png?raw=1)

```python
fig, axs = plt.subplots(2,1)
axs[0].plot(df.month_number , df.bathingsoap, color= 'k', marker = 'o')
axs[0].set_title("BathingSoap")
axs[0].set_xticks([])
axs[1].plot(df.month_number , df.facewash, color = 'r', marker ='o')
axs[1].set_title("FaceWash")
axs[1].set_xlabel(f"Month Number")
axs[1].set_ylabel(f"Sales units in number")
plt.show()
```

### Exercise 10: Read all product sales data and show it using the stack plot

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/10.png?raw=1)

```python
fig, ax = plt.subplots()

# Ly cc nhn ct
labels = df.columns[:-2]
colors = ['red', 'blue', 'green', 'purple', 'orange', 'gray']

x = df.index
y1 = df.facecream
y2 = df.facewash
y3 = df.toothpaste
y4 = df.bathingsoap
y5 = df.shampoo
y6 = df.moisturizer

plt.stackplot(x, y1, y2, y3, y4, y5, y6, labels=['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo','moisturizer'], colors = colors)

ax.set_title('Biu  ng chng')
ax.set_xlabel('Thng')
ax.set_ylabel('Gi tr')

# t ch thch cho biu 
ax.legend(loc='upper left')

# Hin th biu 
plt.show()
```
